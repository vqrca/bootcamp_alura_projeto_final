```python
#Fun칞칚o para plotar gr치fico do tipo countplot

def count_plot(dados, ylabel='', xlabel='', title='', label0='', label1=''):
               
  sns.set_style('white')
  fig, ax = plt.subplots(figsize=(8,5))
  fig.subplots_adjust(left=0.28, right=0.53, top=0.61, bottom=0.107)
  sns.countplot(dados, palette='Paired')
  plt.tight_layout()

  plt.ylabel(ylabel, fontdict = {'fontsize': 18}, color = GRAY3)
  plt.xlabel(xlabel, fontdict = {'fontsize': 20}, color = GRAY3)
  plt.title(title,  loc = 'left', fontsize = 25, color = GRAY4)
  ax.tick_params(color = 'darkgrey', bottom = 'off')
  ax.spines['bottom'].set_color('darkgrey')
  ax.spines['left'].set_color('darkgrey')
  ax.spines['right'].set_visible(False)
  ax.spines['top'].set_visible(False)

  #colocando valor nas barras
  for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
    height + 3,
    '{:1.0f}'.format((height)),
    ha='center', color = GRAY3, fontsize=12)

  fig.canvas.draw()

  labels = [item.get_text() for item in ax.get_xticklabels()]
  labels[0] = label0
  labels[1] = label1

  ax.set_xticklabels(labels)

  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(20)
    i.set_color(GRAY3)
  for i in ax.get_yticklabels() + ax.get_xticklabels():
    i.set_fontsize(20)
    i.set_color(GRAY3)
  plt.show()
```
